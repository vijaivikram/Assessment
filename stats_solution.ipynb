What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook

Solution : 
The correlation represents how strongly two variables are releated. The correlation value ranges from -1 to +1.If the value is towards +1, it  indicates positive correlation and if the value is towards -1, it indicates negative correlation. If the value is zero, then it indicates there is no linear relationship between two variables.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'X': np.random.rand(50) * 10,
        'Y': np.random.rand(50) * 15}

df = pd.DataFrame(data)
corr = df.corr()

sns.heatmap(corr,annot=True)
```
